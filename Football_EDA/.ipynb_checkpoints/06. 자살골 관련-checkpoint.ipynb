### 득점상황에 따른 자살골과 승률(0대0 상황에서 자살골을 허용했을때 패배 확률, 또는 승리 확률)

```python
sql = """SELECT * FROM game_records"""
game_records = db_conn.select_query(sql)
game_records_pd = pd.DataFrame(game_records, columns=['game_id', 'home_team_id', 'away_team_id', 'home_team_score', 'away_team_score', 'winning_team'])
game_records_pd.head()
```

```python
goal_records_temp = pd.DataFrame(goal_records, columns=['match_id', 'id', 'score_team_id', 'own_goal', 'lost_team', 'half_type', 'play_time'])
goal_trace = pd.merge(goal_records_temp, game_records_pd[['game_id', 'home_team_id', 'away_team_id', 'winning_team']], how='left', left_on='match_id', right_on='game_id')
goal_trace = pd.DataFrame(goal_trace, columns=['match_id', 'id', 'score_team_id', 'own_goal', 'lost_team', 'home_team_id', 'away_team_id', 'winning_team', 'home_team_score', 'away_team_score', 'location', 'score_line', 'half_type', 'play_time', 'time_range'])
goal_trace.head()
```

```python
goal_trace_pd = copy.deepcopy(goal_trace)
goal_trace_pd.home_team_score = np.where(goal_trace_pd.home_team_id == goal_trace_pd.score_team_id, 1, 0)
goal_trace_pd.away_team_score = np.where(goal_trace_pd.away_team_id == goal_trace_pd.score_team_id, 1, 0)
goal_trace_pd.home_team_score = goal_trace_pd.groupby(['match_id'])['home_team_score'].cumsum()
goal_trace_pd.away_team_score = goal_trace_pd.groupby(['match_id'])['away_team_score'].cumsum()
goal_trace_pd.score_line = goal_trace_pd.home_team_score.map(str) + ':' + goal_trace_pd.away_team_score.map(str)
goal_trace_pd.time_range = (goal_trace_pd.play_time / 15).astype(int) + 1
goal_trace_pd.time_range = np.where(goal_trace_pd.time_range > 3, 3, goal_trace_pd.time_range)
goal_trace_pd.time_range = np.where(goal_trace_pd.half_type == 2, goal_trace_pd.time_range + 3, goal_trace_pd.time_range)
goal_trace_pd.location = np.where(goal_trace_pd.home_team_id == goal_trace_pd.score_team_id, 1, 0)
own_goal_trace_pd = goal_trace_pd[goal_trace_pd.own_goal == 'Y']

own_goal_trace_pd.head()
```

```python
own_goal_timing_pd = pd.concat([own_goal_trace_pd[own_goal_trace_pd.score_line == '1:0'], own_goal_trace_pd[own_goal_trace_pd.score_line == '0:1']])[['match_id', 'winning_team', 'score_team_id', 'lost_team']]

# 자살골을 넣은 선수가 속한 팀은 lost_team

own_goal_timing_pd = pd.DataFrame(own_goal_timing_pd, columns=['match_id', 'winning_team', 'score_team_id', 'lost_team', 'own_goal_winning_flag'])
own_goal_timing_pd.lost_team = own_goal_timing_pd.lost_team.apply(pd.to_numeric)
own_goal_timing_pd.own_goal_winning_flag = np.where(own_goal_timing_pd.winning_team == 0, 0, np.where(own_goal_timing_pd.winning_team == own_goal_timing_pd.lost_team, 1, -1))
own_goal_timing_pd.__len__(), own_goal_timing_pd[own_goal_timing_pd.own_goal_winning_flag == 1].__len__()

own_goal_timing_statics = pd.DataFrame({'result': ['win', 'draw', 'lose', 'total'], 'idx': [0, 0, 0, 0], 'count': [own_goal_timing_pd[own_goal_timing_pd.own_goal_winning_flag == 1].__len__(), own_goal_timing_pd[own_goal_timing_pd.own_goal_winning_flag == 0].__len__(), own_goal_timing_pd[own_goal_timing_pd.own_goal_winning_flag == -1].__len__(), own_goal_timing_pd.__len__()]})
own_goal_timing_statics = own_goal_timing_statics.pivot(index = 'idx', columns = 'result', values = 'count')
own_goal_timing_statics = own_goal_timing_statics.reset_index()
own_goal_timing_statics = own_goal_timing_statics.drop(['idx'], axis=1)
own_goal_timing_statics.index = ['count']
own_goal_timing_statics

[own_goal_timing_statics.win / own_goal_timing_statics.total, own_goal_timing_statics.draw / own_goal_timing_statics.total, own_goal_timing_statics.lose / own_goal_timing_statics.total, 1.]

own_goal_timing_statics = own_goal_timing_statics.append(pd.DataFrame({'win': own_goal_timing_statics.win / own_goal_timing_statics.total, 'draw': own_goal_timing_statics.draw / own_goal_timing_statics.total, 'lose': own_goal_timing_statics.lose / own_goal_timing_statics.total, 'total': 1.}))
own_goal_timing_statics = own_goal_timing_statics[['win', 'draw', 'lose', 'total']]
own_goal_timing_statics.index = ['count', 'probs']
own_goal_timing_statics
```

#### 결과
* 전체 자살골 109골 중 35골이 선제골 역할을 수행
* 35골이 포함된 경기(35경기) 중 승리 4경기, 무승부 5경기, 패배 26경기
* 거의 대다수인 74.3%가 패배하고, 11.4%만 자살골을 극복하고 승리

### 1대0 앞선 상황에서 동점골을 자살골로 허용햇을때의 패배, 또는 승리 확률

```python
own_tying_goal = pd.merge(goal_records_pd[goal_records_pd.own_goal == 'Y'], game_records_pd, how='left', left_on='match_id', right_on='game_id')
own_tying_goal = pd.DataFrame(own_tying_goal[['match_id', 'id', 'winning_team', 'lost_team']], columns = ['match_id', 'id', 'winning_team', 'lost_team', 'previous_goal_id'])

own_tying_goal.previous_goal_id = own_tying_goal.id - 1
own_tying_goal = own_tying_goal.reset_index(drop=True)

previous_goal = copy.deepcopy(goal_records_pd[['match_id', 'id', 'score_team_id', 'lost_team']])
previous_goal = pd.merge(previous_goal, game_records_pd, how='left', left_on='match_id', right_on='game_id')
previous_goal = pd.DataFrame(previous_goal, columns=['match_id', 'id', 'score_team_id', 'lost_team', 'home_team_id', 'away_team_id', 'home_score', 'away_score', 'score_line'])
previous_goal.home_score = np.where(previous_goal.score_team_id == previous_goal.home_team_id, 1, 0)
previous_goal.away_score = np.where(previous_goal.score_team_id == previous_goal.away_team_id, 1, 0)
previous_goal.home_score = previous_goal.groupby(['match_id']).agg({'home_score': 'cumsum'})
previous_goal.away_score = previous_goal.groupby(['match_id']).agg({'away_score': 'cumsum'})
previous_goal.score_line = previous_goal.home_score.map(str) + ':' + previous_goal.away_score.map(str)

own_tying_goal = pd.merge(own_tying_goal, previous_goal.loc[own_tying_goal.previous_goal_id-2, ['id', 'home_team_id', 'away_team_id', 'home_score', 'away_score', 'score_line']].reset_index(drop=True), how='left', left_on='previous_goal_id', right_on='id')

own_tying_goal = pd.DataFrame(own_tying_goal[['match_id', 'id_x', 'winning_team', 'lost_team', 'home_team_id', 'away_team_id', 'home_score', 'away_score', 'score_line']], columns = ['match_id', 'id_x', 'winning_team', 'lost_team', 'home_team_id', 'away_team_id', 'home_score', 'away_score', 'score_line', 'previous_status', 'winning_flag'])

own_tying_goal.lost_team = own_tying_goal.lost_team.apply(pd.to_numeric)

own_tying_goal.winning_flag = np.where(own_tying_goal.winning_team == 0, 0, np.where(own_tying_goal.winning_team == own_tying_goal.lost_team, 1, -1))

own_tying_goal.head()
```

```python
own_tying_goal.previous_status = np.where(own_tying_goal.lost_team == own_tying_goal.home_team_id, own_tying_goal.home_score-own_tying_goal.away_score, own_tying_goal.away_score-own_tying_goal.home_score)

own_tying_goal_pd = own_tying_goal[own_tying_goal.previous_status == 1]
own_tying_goal_pd.columns = ['game_id', 'id', 'winning_team', 'lost_team', 'home_team_id', 'away_team_id', 'pre_home_score', 'pre_away_score', 'pre_score_line', 'previous_diff', 'winning_flag']
own_tying_goal_pd = pd.merge(own_tying_goal_pd, game_records_pd[['game_id', 'home_team_score', 'away_team_score']], how='left', on='game_id')
own_tying_goal_pd.head()
```

```python
own_tying_goal_statics = pd.DataFrame({'result': ['win', 'draw', 'lose', 'total'], 'idx': [0, 0, 0, 0], 'count': [own_tying_goal_pd[own_tying_goal_pd.winning_flag == 1].__len__(), own_tying_goal_pd[own_tying_goal_pd.winning_flag == 0].__len__(), own_tying_goal_pd[own_tying_goal_pd.winning_flag == -1].__len__(), own_tying_goal_pd.__len__()]})

own_tying_goal_statics = own_tying_goal_statics.pivot(index = 'idx', columns = 'result', values = 'count')
own_tying_goal_statics = own_tying_goal_statics.reset_index()
own_tying_goal_statics = own_tying_goal_statics.drop(['idx'], axis=1)
own_tying_goal_statics.index = ['count']
own_tying_goal_statics

[own_tying_goal_statics.win / own_tying_goal_statics.total, own_tying_goal_statics.draw / own_tying_goal_statics.total, own_tying_goal_statics.lose / own_tying_goal_statics.total, 1.]

own_tying_goal_statics = own_tying_goal_statics.append(pd.DataFrame({'win': own_tying_goal_statics.win / own_tying_goal_statics.total, 'draw': own_tying_goal_statics.draw / own_tying_goal_statics.total, 'lose': own_tying_goal_statics.lose / own_tying_goal_statics.total, 'total': 1.}))
own_tying_goal_statics = own_tying_goal_statics[['win', 'draw', 'lose', 'total']]
own_tying_goal_statics.index = ['count', 'probs']
own_tying_goal_statics
```

#### 결과
