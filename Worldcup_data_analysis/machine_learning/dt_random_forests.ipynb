```python
import pandas as pd
import db_conn

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
src = pd.read_csv('./embeded_ps.csv', sep='\t')
src = src.iloc[:, 1:]
```

```python
src.head()
```

```python
src_wo_draw = src[src.match_id.apply(lambda x: x not in [3,7,9,22,32,35,36,37,42,51,52,56,59])]
src_wo_draw.head()
```

```python
team_info = pd.DataFrame(db_conn.select_query('select * from team'))
player_info = pd.DataFrame(db_conn.select_query('select * from player'))
match_info = pd.DataFrame(db_conn.select_query('select * from wc_match'))
```

```python
src_wo_draw['team_id'] = src_wo_draw.player_id.apply(lambda x: player_info[player_info.id == x].team_id.values[0])
src_wo_draw.head()
```

```python
match_info['winning_team_id'] = match_info.apply(lambda x: x.home_team_id if x.home_team_score > x.away_team_score else x.away_team_id, axis=1)
```

```python
match_info.head()
```

```python
src_wo_draw['winning_flag'] = src_wo_draw.apply(lambda x: 1 if match_info[(match_info.id == x.match_id) & (match_info.winning_team_id == x.team_id)].__len__() > 0 else 0, axis=1)
src_wo_draw.head()
```
