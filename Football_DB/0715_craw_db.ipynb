```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# overall list
ENTRANCE_URL = 'https://www.w3resource.com/sql-exercises/soccer-database-exercise/'
req = requests.get('https://www.w3resource.com/sql-exercises/soccer-database-exercise/index.php')
html = req.text

tables = {}
if req.ok:
    soup = BeautifulSoup(html, 'html.parser')
    table_ul = soup.select('article > ul')[0]
    table_as = table_ul.select('li > a')
    
    for i in range(len(table_as)):
        db_name = table_as[i].text
        db_loc = table_as[i].get('href')
        tables[db_name] = {'db_loc':db_loc}
```

```python
tables
```

```python
# columns
# columns are in order
tables['soccer_country']['columns'] = ['country_id', 'country_abbr', 'country_name']
tables['soccer_city']['columns'] = ['city_id', 'city_abbr', 'country_id']
tables['soccer_venue']['columns'] = ['venue_id', 'venue_name', 'city_id', 'aud_capacity']
tables['soccer_team']['columns'] = ['team_id', 'team_group', 'match_played', 'won', 'draw', 'lost', 'goal_for', 'goal_agnst', 'goal_diff', 'points', 'group_position']
tables['playing_position']['columns'] = ['position_id', 'position_desc']
tables['player_mast']['columns'] = ['player_id', 'team_id', 'jersey_no', 'player_name', 'posi_to_play', 'dt_of_bir', 'age', 'playing_club']
tables['referee_mast']['columns'] = ['referee_id', 'referee_name', 'country_id']
tables['match_mast']['columns'] = ['match_no', 'play_stage', 'play_date', 'results', 'decided_by', 'goal_score', 'venue_id', 'referee_id', 'audence', 'plr_of_match', 'stop1_sec', 'stop2_sec']
tables['coach_mast']['columns'] = ['coach_id', 'coach_name']
tables['asst_referee_mast']['columns'] = ['ass_ref_id', 'ass_ref_name', 'country_id']
tables['match_details']['columns'] = ['match_no', 'play_stage', 'team_id', 'win_lose', 'decided_by', 'goal_score', 'penalty_score', 'ass_ref', 'player_gk']
tables['goal_details']['columns'] = ['goal_id', 'match_no', 'player_id', 'team_id', 'goal_time', 'goal_type', 'play_stage', 'goal_schedule', 'goal_half']

tables['penalty_shootout']['columns'] = ['kick_id', 'match_no', 'team_id', 'player_id', 'score_goal', 'kick_no']
tables['player_booked']['columns'] = ['match_no', 'team_id', 'player_id', 'booking_time', 'sent_off', 'play_schedule', 'play_half']
tables['player_in_out']['columns'] = ['match_no', 'team_id', 'player_id', 'in_out', 'time_in_out', 'play_schedule', 'play_half']
tables['match_captain']['columns'] = ['match_no', 'team_id', 'player_captain']
tables['team_coaches']['columns'] = ['team_id', 'coach_id']
tables['penalty_gk']['columns'] = ['match_no', 'team_id', 'player_gk']
```

```python
for k, v in tables.items():
    print(k)
    req = requests.get(ENTRANCE_URL+v['db_loc'])
    html = req.text
#     print(html)

    columns = v['columns']
    print(columns)
    tab_list = []
    if req.ok:
        soup = BeautifulSoup(html, 'html.parser')
        tab_content = soup.select('table > tbody > tr')

        for i in range(1, len(tab_content)):
            row = tab_content[i].select('td')
    #         print(row)
            ta_dict = {}
            for rr in range(len(row)):
                ta_dict[columns[rr]] = row[rr].text
            tab_list.append(ta_dict)
    
    df = pd.DataFrame(tab_list)
    print(df.head())
    df.to_csv('./dataset/{}.csv'.format(k), sep=',', header=True, index=False)
```
